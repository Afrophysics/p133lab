```python
import numpy as np
arch = np.array([220, 208, 187, 199, 235, 231, 150, 166, 212])

#mean
mean = 0
for x in arch:
    mean += x/len(arch)
print('meann: %.3f'%mean)
#sample variance
mean2 = 0
for x in arch:
    val = (x- mean)**2
    mean2 += val/(len(arch) - 1)

print('Variance: %.3f'%mean2)
#sample standard deviation
stndrd_dev = mean2**0.5
print('Standard deviation: %.3f'%stndrd_dev)

#sample mean standard deviation
smple_sd = (mean2/len(arch))**0.5
print('mean standard deviation: %.3f'%smple_sd)
```

```python
import numpy as np

data = np.loadtxt('Statistics_homework_prob7_data.txt')
cur = np.empty(len(data))
vol = np.empty(len(data))
s_v = np.empty(len(data))

for n in range(len(data)):
    cur[n], vol[n], s_v[n] = data[n]
    

cur1 = np.empty(len(data) - 5)
vol1 = np.empty(len(data) - 5)
s_v1 = np.empty(len(data) - 5)

for n in range(len(data) - 5):
    cur1[n], vol1[n], s_v1[n] = data[n]
    
print("Part (a) is meant to build the algorithm")

#Expectation value of the nth-power of the x-array weighted by the inverse of the y-variance
def u(x, sig, n):
    tp = x**n
    bt = sig**2
    ary = tp/bt
    return np.sum(ary)

#Expectation value of the inner product between the the nth-power of the x-array and the y-array weighted by the inverse of the y-variance
def w(x, y, sig, n):
    tp = y*(x**n)
    bt = sig**2
    ary = tp/bt
    return np.sum(ary)
    
#Linear-fit algorithm using least-squares method
def linfit(x_data, y_data, y_error):
    u_0 = u(x_data, y_error, 0)
    u_1 = u(x_data, y_error, 1)
    u_2 = u(x_data, y_error, 2)
    
    w_0 = w(x_data, y_data, y_error, 0)
    w_1 = w(x_data, y_data, y_error, 1)
    w_2 = w(x_data, y_data, y_error, 2)
    
    D = (u_0 * u_2) - (u_1 ** 2)
    stnd_1 = (u_0 * w_1) - (w_0 * u_1)
    stnd_2 = (u_2 * w_0) - (w_1 * u_1)
    
    slope = stnd_1/D
    var_sl = u_0/D
    
    y_intr = stnd_2/D
    var_yt = u_2/D
    
    return np.array([slope, y_intr, var_sl, var_yt])
    
print("Part (a) complete")
    
    
print("Part (b) is shall test the algorithm using the sample data below:")

x = cur
y = vol
yerr = s_v

print('x data points: %s'%x)
print('y data points: %s'%y)
print('y-error data points: %s'%yerr)

rs = linfit(x, y, yerr)
print("results")
print("slope = %.3f +/- %.3f"%(rs[0], rs[2]**0.5))
print("intercept = %.3f +/- %.3f"%(rs[1], rs[3]**0.5))

print("Part (b) complete")
print('')
print("Part (c) we shall plot our results")
import matplotlib.pyplot as plt

fig = plt.figure()
plt.errorbar(x, y, yerr = yerr, fmt='.', label='resistor potential')
smpl_x = np.linspace(0, 1, 41)

f_x = rs[1] + (rs[0] * smpl_x)
plt.plot(smpl_x, f_x, '--', label='Linear-fit')
plt.grid(True)
plt.title('I-V Diagram')
plt.legend()
plt.xlabel('Current (Amps)')
plt.ylabel('Potential (V)')
plt.show()
fig.savefig('I_V_hw3_ex.png')

print("Problem 2 complete")



x = cur1
y = vol1
yerr = s_v1

print('x data points: %s'%x)
print('y data points: %s'%y)
print('y-error data points: %s'%yerr)

rs1 = linfit(x, y, yerr)
print("results")
print("slope = %.3f +/- %.3f"%(rs[0], rs[2]**0.5))
print("intercept = %.3f +/- %.3f"%(rs[1], rs[3]**0.5))

print("Part (b) complete")
print('')
print("Part (c) we shall plot our results")
import matplotlib.pyplot as plt

fig = plt.figure()
plt.errorbar(x, y, yerr = yerr, fmt='.', label='resistor potential')
smpl_x = np.linspace(0, 1, 41)

f_x = rs1[1] + (rs1[0] * smpl_x)
plt.plot(smpl_x, f_x, '--', label='Linear-fit')
plt.grid(True)
plt.title('I-V Diagram')
plt.legend()
plt.xlabel('Current (Amps)')
plt.ylabel('Potential (V)')
plt.show()
fig.savefig('I_V_hw3_exfix.png')

print("Problem 2 complete")
```
